You are assigned to create a view that only shows the most updated transactions for each id, so there is no duplicate id inside the view.

#### Import Library

```python
import sqlite3
import pandas as pd
```

#### Create Function

```python
def db_query(record,query,filename):
    conn=sqlite3.connect(filename) ##create connection to database and 
    c = conn.cursor()
    c.execute('''CREATE TABLE transactions
                   (id INTEGER, insert_time DATETIME, tx_amout FLOAT, tx_type TEXT, status TEXT)''') ## create table schema
    c.executemany("INSERT INTO transactions VALUES (?,?,?,?,?)",record) #insert record to table
    print('Success create database')
    c.execute(query)
    print('Query success')
    items=c.fetchall()
    for item in items:
        print (item)
    conn.close()
```

#### Execute Query Function

### Question 2
Using Dataset B above, Write a query to get the rolling sum total of sales using the window function!

```python
def db_query_sum(record,query,filename):
    conn=sqlite3.connect(filename) ##create connection to database and 
    c = conn.cursor()
    c.execute('''CREATE TABLE sales
                   (Date DATE, Sales FLOAT)''') ## create table schema
    c.executemany("INSERT INTO sales VALUES (?,?)",record) #insert record to table
    print('Success create database')
    c.execute(query)
    print('Query success')
    items=c.fetchall()
    for item in items:
        print (item)
    conn.close()
```
